```python
import pandas as pd
import json
import plotly
```

```python
with open('shopee_raw.json', 'r') as f:
    shopee = json.load(f)
```

```python
Id = []
Produk = []
Stok = []
Lokasi = []
Terjual = []
Harga = []
Diskon = []
Rating = []
Penilaian = []
Like = []
Verified = []
```

```python
for item in shopee['items']:
    print(f'''- ID        : {item['item_basic']['itemid']}
- Product   : {item['item_basic']['name']}
- Stok      : {item['item_basic']['stock']}
- Lokasi    : {item['item_basic']['shop_location']}
- Terjual   : {item['item_basic']['historical_sold']}
- Harga     : {int(item['item_basic']['price_min']/100000)} - {int(item['item_basic']['price_max']/100000)}
- Diskon    : {item['item_basic']['discount']}
- Rating    : {"%.2f" % item['item_basic']['item_rating']['rating_star']}
- Penilaian : {item['item_basic']['item_rating']['rating_count'][0]}
- Like      : {item['item_basic']['liked_count']}
- Verified  : {'Ya' if item['item_basic']['shopee_verified'] else 'Tidak'}
------------------------------------------------------------------
''')
    Id.append(item['item_basic']['itemid'])
    Produk.append(item['item_basic']['name'])
    Stok.append(item['item_basic']['stock'])
    Lokasi.append(item['item_basic']['shop_location'])
    Terjual.append(item['item_basic']['historical_sold'])
    Harga.append(f'''{int(item['item_basic']['price_min']/100000)} - {int(item['item_basic']['price_max']/100000)}''')
    Diskon.append(item['item_basic']['discount'])
    Rating.append(round(item['item_basic']['item_rating']['rating_star'], 2))
    Penilaian.append(item['item_basic']['item_rating']['rating_count'][0])
    Like.append(item['item_basic']['liked_count'])
    Verified.append(item['item_basic']['shopee_verified'])
```

```python
df_shopee = pd.DataFrame(list(zip(Id, Produk, Harga, Lokasi, Stok, Terjual, Diskon, Rating, Penilaian, Like, Verified)), columns=['Id', 'Produk', 'Harga', 'Lokasi', 'Stok', 'Terjual', 'Diskon', 'Rating', 'Penilaian', 'Like', 'Verified'])
df_shopee.head()
```
