```python
import pandas as pd
import json
import numpy as np
import plotly
import os
```

```python
with open('DataShopee/shopee_gramedia_raw_1.json', 'r') as f:
    data = json.load(f)
data = data['data']['sections'][0]['data']['item']
```

```python
Id = []
Produk = []
Stok = []
Lokasi = []
Terjual = []
Harga = []
Diskon = []
Rating = []
Penilaian = []
Like = []
Verified = []
Brand = []
```

```python
for page in os.listdir('DataShopee'):
    with open(f'DataShopee/{page}', 'r') as f:
        data = json.load(f)
    data = data['data']['sections'][0]['data']['item']          
    for i in data:
        Id.append(i['itemid'])
        Produk.append(i['name'])
        Stok.append(i['stock'])
        Lokasi.append(i['shop_location'])
        Terjual.append(i['historical_sold'])
        Brand.append(i['brand'])
        Harga.append(f'''{int(i['price_min']/100000)} - {int(i['price_max']/100000)}''')
        Diskon.append(i['discount'])
        Rating.append(round(i['item_rating']['rating_star'], 2))
        Penilaian.append(i['item_rating']['rating_count'][0])
        Like.append(i['liked_count'])
```

```python
shopee = pd.DataFrame(list(zip(Id, Produk, Brand, Lokasi, Harga, Stok, Terjual, Diskon, Rating, Penilaian, Like)), columns=['Id', 'Produk', 'Brand', 'Lokasi', 'Harga', 'Stok', 'Terjual', 'Diskon', 'Rating', 'Penilaian', 'Like'])
shopee.head()
```
